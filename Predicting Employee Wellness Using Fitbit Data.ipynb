```python
# Feature Engineering

# Calculate BMI
df['BMI'] = df['weight'] / ((df['height'] / 100) ** 2)

# Define mental wellness conditions with adjusted thresholds
def determine_mental_wellness(row):
    score = 0
    if row['steps'] >= 3000:
        score += 1
    if row['resting_heart'] <= 85:
        score += 1
    if row['entropy_heart'] >= threshold_entropy:
        score += 1
    if row['calories'] >= 1500:
        score += 1
    if row['BMI'] <= 27:
        score += 1
    if 'Lying' not in row['activity'] and 'Sitting' not in row['activity']:
        score += 1
        
    return 1 if score >= 4 else 0  # Adjust threshold score as needed

# Set a threshold for entropy_heart based on analysis of your data
threshold_entropy = df['entropy_heart'].mean() * 0.8

# Apply the function to create the target column
df['mental_wellness'] = df.apply(determine_mental_wellness, axis=1)
```

```python
X=df.drop(columns='mental_wellness')
y=df['mental_wellness']
```
