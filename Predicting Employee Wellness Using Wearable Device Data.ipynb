#  For predicting wellness score we need extra feature from these features and then calculate wellness score

```python
# Activity Intensity
df['activity_intensity'] = pd.cut(df['steps'], bins=[0, 5000, 10000, float('inf')], labels=['Low', 'Moderate', 'High'])
```

```python
df.head(2)
```

# Calculating BMI (BMI = weight/(height/100)^2)

```python

df['BMI']=df['weight']/(df['height']/100)**2
# healthy bmi ranges from 18.5 to 24.9
```

```python
df.head()
```

#  Calculating Wellness score

```python
df['wellness_score']=(df['steps']*0.3)+((70-df['resting_heart'])*0.2)+((24.9-df['BMI'].clip(lower=0)*0.2)+(df['intensity_karvonen']*0.3))
```

```python
new_df=df[['steps','hear_rate','resting_heart','BMI','intensity_karvonen','wellness_score']]
```

```python
X=new_df.drop(columns='wellness_score')
y=new_df['wellness_score']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=42,test_size=0.3)
```

```python
X_train
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr_model=LinearRegression()
```

```python
lr_model.fit(X_train,y_train)
```

```python
y_pred=lr_model.predict(X_test)
```

```python
from sklearn.metrics import r2_score,root_mean_squared_error,mean_squared_error
```

```python
mse=mean_squared_error(y_test,y_pred)
print(f'Mean squared error is: {mse:.30f}')
rmse=root_mean_squared_error(y_test,y_pred)
print(f'Root Mean squared error is:{rmse:.20f}')
r2score=r2_score(y_test,y_pred)
print('R2 score is:',r2score)
```

# Lets check the r2 score on training data 

```python
X_train_prediction=lr_model.predict(X_train)
training_data_r2score=r2_score(X_train_prediction,y_train)
training_data_r2score
```
